```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import pybamm
import optuna
from optuna.samplers import TPESampler

optuna.logging.set_verbosity(optuna.logging.WARNING)
```

```python
# ============================================================
# OKane2022 OPTUNA FITTER
# Uses Bayesian optimisation (TPE sampler) to minimise RMSE
# between simulated and measured discharge voltage.
# ============================================================

from concurrent.futures import ProcessPoolExecutor, TimeoutError as FuturesTimeoutError

optuna.logging.set_verbosity(optuna.logging.WARNING)

# ============================================================
# SETTINGS
# ============================================================

CSV_FILE            = "Trace_2025-03-27_3.csv"
INITIAL_SOC         = 0.9811111
DISCHARGE_CURRENT_A = 2
DISCHARGE_TIME_MIN  = 20

N_TRIALS            = 200      # total Optuna trials to run
SCALE_LOW           = 0.7      # lower bound: fraction of OKane2022 default
SCALE_HIGH          = 1.3      # upper bound: fraction of OKane2022 default
SOLVE_TIMEOUT_S     = 10      # seconds before a hanging solver subprocess is killed
MIN_SIM_COVERAGE    = 0.8      # prune if sim covers less than this fraction of the discharge window

# ============================================================
# Params to optimise.
# Each is sampled as a continuous uniform float in
# [default * SCALE_LOW, default * SCALE_HIGH].
#
# DO NOT add here:
#   - integer params (number of electrodes)
#   - voltage cutoffs
#   - function-valued params (OCV curves, diffusivities)
#   - anything with a default of 0
# ============================================================

PARAMS_TO_OPTIMISE = [
    "Positive electrode thickness [m]",
    "Negative electrode thickness [m]",
    "Electrode height [m]",
    "Electrode width [m]",
    "Negative electrode porosity",
    "Positive electrode porosity",
    "Initial concentration in electrolyte [mol.m-3]"
]

# Neg + pos electrode concentrations swept as one coupled axis
# so their ratio (stoichiometry) is preserved.
OPTIMISE_ELECTRODE_CONC = True

# ============================================================
# Params that override OKane2022 defaults for every trial.
# ============================================================

FIXED_OVERRIDES = {
    "Nominal cell capacity [A.h]":          3.5,
    "Lower voltage cut-off [V]":            0.0,
    "Upper voltage cut-off [V]":            3.8,
    "Open-circuit voltage at 0% SOC [V]":   0.0,
    "Open-circuit voltage at 100% SOC [V]": 3.6,
}

# ============================================================
# Load OKane2022 defaults once
# ============================================================

_DEFAULTS = pybamm.ParameterValues("OKane2022")

def default(name):
    val = _DEFAULTS[name]
    if callable(val):
        raise ValueError(f"'{name}' is function-valued and cannot be scaled directly.")
    return float(val)

PARAM_BASES = {name: default(name) for name in PARAMS_TO_OPTIMISE}
if OPTIMISE_ELECTRODE_CONC:
    CONC_NEG_BASE = default("Initial concentration in negative electrode [mol.m-3]")
    CONC_POS_BASE = default("Initial concentration in positive electrode [mol.m-3]")

print("OKane2022 base values:")
for name, val in PARAM_BASES.items():
    print(f"  {name}: {val:.6g}  -->  [{val*SCALE_LOW:.6g}, {val*SCALE_HIGH:.6g}]")
if OPTIMISE_ELECTRODE_CONC:
    print(f"  Electrode conc (neg, coupled): {CONC_NEG_BASE:.6g}  -->  [{CONC_NEG_BASE*SCALE_LOW:.6g}, {CONC_NEG_BASE*SCALE_HIGH:.6g}]")
    print(f"  Electrode conc (pos, coupled): {CONC_POS_BASE:.6g}  -->  [{CONC_POS_BASE*SCALE_LOW:.6g}, {CONC_POS_BASE*SCALE_HIGH:.6g}]")
print()

# ============================================================
# Helpers
# ============================================================

def load_real_voltage_data(csv_file):
    df = pd.read_csv(csv_file, skiprows=8)
    df.columns = ["Time", "Voltage", "Power"]
    df["Time"] = pd.to_datetime(df["Time"])
    df["Time (s)"] = (df["Time"] - df["Time"].iloc[0]).dt.total_seconds()

    voltage_diff   = df["Voltage"].diff()
    drop_time      = df.loc[voltage_diff.idxmin(), "Time (s)"]
    rise_time      = df.loc[voltage_diff.idxmax(), "Time (s)"]
    drop_magnitude = abs(voltage_diff.min())
    df.loc[
        (df["Time (s)"] >= drop_time) & (df["Time (s)"] < rise_time),
        "Voltage"
    ] += drop_magnitude

    return df["Time (s)"].to_numpy(), df["Voltage"].to_numpy()



def _solve_worker(params_dict, conc_neg, conc_pos, fixed_overrides,
                  discharge_a, discharge_min, initial_soc):
    """Runs in a subprocess — no shared state with the main process."""
    import pybamm
    model = pybamm.lithium_ion.DFN({
        "thermal": "lumped",
        "lithium plating": "irreversible",
        "lithium plating porosity change": "true",
    })
    params = pybamm.ParameterValues("OKane2022")
    params.update(fixed_overrides, check_already_exists=False)
    if conc_neg is not None:
        params.update({
            "Initial concentration in negative electrode [mol.m-3]": conc_neg,
            "Initial concentration in positive electrode [mol.m-3]": conc_pos,
        }, check_already_exists=False)
    params.update(params_dict, check_already_exists=False)
    experiment = pybamm.Experiment([
        (f"Discharge at {discharge_a} A for {discharge_min} minutes",)
    ])
    sim = pybamm.Simulation(model, parameter_values=params, experiment=experiment)
    solution = sim.solve(initial_soc=initial_soc)
    return (solution["Time [s]"].entries, solution["Voltage [V]"].entries)


def run_simulation(param_dict: dict, conc_scale: float, real_time_s: np.ndarray):
    conc_neg = CONC_NEG_BASE * conc_scale if OPTIMISE_ELECTRODE_CONC else None
    conc_pos = CONC_POS_BASE * conc_scale if OPTIMISE_ELECTRODE_CONC else None

    with ProcessPoolExecutor(max_workers=1) as executor:
        future = executor.submit(
            _solve_worker,
            param_dict, conc_neg, conc_pos, FIXED_OVERRIDES,
            DISCHARGE_CURRENT_A, DISCHARGE_TIME_MIN, INITIAL_SOC,
        )
        try:
            sim_time_s, sim_voltage_v = future.result(timeout=SOLVE_TIMEOUT_S)
        except FuturesTimeoutError:
            future.cancel()
            raise RuntimeError(f"Solver timed out after {SOLVE_TIMEOUT_S}s")

    expected_s = DISCHARGE_TIME_MIN * 60
    if sim_time_s[-1] < expected_s * MIN_SIM_COVERAGE:
        raise RuntimeError(
            f"Solver only reached {sim_time_s[-1]:.0f}s of {expected_s}s "
            f"({100*sim_time_s[-1]/expected_s:.0f}% coverage)"
        )

    valid          = real_time_s <= sim_time_s[-1]
    voltage_interp = np.interp(real_time_s[valid], sim_time_s, sim_voltage_v)
    return real_time_s[valid], voltage_interp


def rmse(y_true, y_pred):
    return np.sqrt(np.mean((y_true - y_pred) ** 2))


# ============================================================
# Optuna objective
# ============================================================

real_time_s, real_voltage_v = load_real_voltage_data(CSV_FILE)

def objective(trial):
    # Sample each param as a uniform float in [base * SCALE_LOW, base * SCALE_HIGH]
    param_dict = {
        name: trial.suggest_float(name, base * SCALE_LOW, base * SCALE_HIGH)
        for name, base in PARAM_BASES.items()
    }

    conc_scale = 1.0
    if OPTIMISE_ELECTRODE_CONC:
        conc_scale = trial.suggest_float(
            "electrode_conc_scale", SCALE_LOW, SCALE_HIGH
        )

    try:
        fit_time_s, sim_voltage_v = run_simulation(param_dict, conc_scale, real_time_s)
        error = rmse(real_voltage_v[:len(fit_time_s)], sim_voltage_v)
        return error
    except Exception as e:
        # Tell Optuna this trial failed; it will avoid this region
        raise optuna.exceptions.TrialPruned() from e


# ============================================================
# Run optimisation
# ============================================================

sampler = TPESampler(seed=42)
study   = optuna.create_study(direction="minimize", sampler=sampler)

print(f"Running {N_TRIALS} Optuna trials...\n")

completed = 0
pruned    = 0

def trial_callback(study, trial):
    global completed, pruned
    if trial.state == optuna.trial.TrialState.COMPLETE:
        completed += 1
        print(f"  Trial {trial.number:>4}  RMSE = {trial.value:.6g} V"
              + ("  [best]" if trial.number == study.best_trial.number else ""))
    elif trial.state == optuna.trial.TrialState.PRUNED:
        pruned += 1
        print(f"  Trial {trial.number:>4}  FAILED (pruned)")

study.optimize(objective, n_trials=N_TRIALS, callbacks=[trial_callback])

# ============================================================
# Report best result
# ============================================================

best = study.best_trial
print(f"\n================ BEST FIT ================")
print(f"Trial:     {best.number}")
print(f"RMSE:      {best.value:.6g} V")
print(f"Completed: {completed}  |  Pruned/failed: {pruned}")
print(f"\nBest parameters:")
for name, val in best.params.items():
    if name == "electrode_conc_scale":
        print(f"  electrode_conc_scale: {val:.6g}")
        print(f"    --> neg conc: {CONC_NEG_BASE * val:.6g}  (default: {CONC_NEG_BASE:.6g})")
        print(f"    --> pos conc: {CONC_POS_BASE * val:.6g}  (default: {CONC_POS_BASE:.6g})")
    else:
        base = PARAM_BASES[name]
        print(f"  {name}: {val:.6g}  (x{val/base:.3f} of default {base:.6g})")
print("==========================================\n")

# ============================================================
# Re-run best params to get voltage trace for plotting
# ============================================================

best_param_dict = {
    name: val for name, val in best.params.items()
    if name != "electrode_conc_scale"
}
best_conc_scale = best.params.get("electrode_conc_scale", 1.0)
best_time_s, best_voltage_v = run_simulation(best_param_dict, best_conc_scale, real_time_s)

# ============================================================
# Plots
# ============================================================

# 1. Optimisation history
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

trials_df = study.trials_dataframe()
completed_df = trials_df[trials_df["state"] == "COMPLETE"]

axes[0].plot(completed_df["number"], completed_df["value"], "o", markersize=3, alpha=0.5)
axes[0].plot(
    completed_df["number"],
    completed_df["value"].cummin(),
    color="red", linewidth=1.5, label="Best so far"
)
axes[0].set_xlabel("Trial")
axes[0].set_ylabel("RMSE (V)")
axes[0].set_title("Optimisation history")
axes[0].legend()
axes[0].grid(True)

# 2. Best fit vs measured
axes[1].plot(real_time_s, real_voltage_v, label="Measured")
axes[1].plot(best_time_s, best_voltage_v, "--",
             label=f"Best fit  RMSE={best.value:.4f} V")
axes[1].set_xlabel("Time (s)")
axes[1].set_ylabel("Voltage (V)")
axes[1].set_title("Best-fit simulation vs measured")
axes[1].legend()
axes[1].grid(True)

plt.tight_layout()
plt.show()

# 3. Parameter importance
try:
    importances = optuna.importance.get_param_importances(study)
    names  = list(importances.keys())
    values = list(importances.values())

    plt.figure(figsize=(8, max(3, len(names) * 0.5)))
    bars = plt.barh(names[::-1], values[::-1])
    plt.xlabel("Importance")
    plt.title("Parameter importance (Optuna FAnova)")
    plt.tight_layout()
    plt.show()
except Exception:
    pass  # importance requires enough completed trials; skip silently if not
```
